```python
# initialize

import parselmouth
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# filepath for the dataset
root = "/Users/leochoo/dev/GP2-dev/SVD"

my_data = root + "/my_data"
healthy = root + "/healthy"
patho = root + "/pathological"

!ls {patho}

# on MacOS, it creates a hidden .DS_Store folder that only be shown using `ls -la` command. You should delete it to ge the code working! :)
```

```python
# functions

def get_jitter_local(_path):
    sounds = os.listdir(_path)
    jitter = []
    for sound in sounds: 
        snd = parselmouth.Sound(_path + "/" + sound)
        # convert the object
        pitch = snd.to_pitch()
        pulses = parselmouth.praat.call([snd, pitch], "To PointProcess (cc)")
        # jitter
        jitter_local = parselmouth.praat.call(pulses, "Get jitter (local)", 0.0, 0.0, 0.0001, 0.02, 1.3) * 100
        jitter.append(jitter_local)
    return jitter

def get_shimmer():
    pass


def get_hnr(): #harmonic to noise ratio
    pass
```

```python
import time
start = time.time()
j_healthy = get_jitter_local(healthy)
end = time.time()
print("healthy: ",end - start)
```

```python
start = time.time()
j_psycho = get_jitter_local(patho + "/psychogenic")
end = time.time()
print("psycho: ",end - start)
```

```python
# graph scatter plot

sns.set()

fig = plt.figure()
plt.title("Healthy vs Patho")
plt.xlabel("n")
plt.ylabel("jitter")

plt.xlim(right=100)
plt.ylim(top=2)
plt.legend(loc=2)

plt.scatter(list(range(len(j_healthy))), j_healthy, c='r')
plt.scatter(list(range(len(j_psycho))), j_psycho, c="b")

plt.show()
```
